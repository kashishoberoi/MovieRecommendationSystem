```python
import pandas as pd
```

```python
movies = pd.read_csv('./../../data/movies.csv')
```

```python
movie_dict = dict()
for itr in range(len(movies)):
    movie_dict[movies.iloc[itr]['movieId']] = dict()
    movie_dict[movies.iloc[itr]['movieId']]['title'] = movies.iloc[itr]['title']
    movie_dict[movies.iloc[itr]['movieId']]['genres'] = movies.iloc[itr]['genres'].split('|')
```

```python
genre_set = []
for gen in movies['genres'].values:
    genre_set.extend(gen.split('|'))
genre_set = set(genre_set)
```

```python
genre_set = list(genre_set - set(['(no genres listed)']))
```

```python
ratings = pd.read_csv('./../../data/ratings.csv')
ratings_movie_mean =  ratings.groupby('movieId')['rating'].mean()
```

```python
gen_dict = dict()
for gen in genre_set:
    gen_dict[gen] = []
    for movie_id in movie_dict.keys():
        if gen in movie_dict[movie_id]['genres']:
            try:
                rating = ratings_movie_mean[movie_id]
            except KeyError as e:
                rating = 0
            gen_dict[gen].append([movie_id,movie_dict[movie_id]['title'],round(rating,2)])
```

```python
gen_list = []
for gen in gen_dict.keys():
    for itr in range(len(gen_dict[gen])):
        if gen_dict[gen][itr][2] >=4.5:
            gen_list.append([gen_dict[gen][itr][0],gen_dict[gen][itr][1],gen,gen_dict[gen][itr][2]])
```

```python
len(gen_list)
```

```python
import sqlite3
```

```python
def connect_movie_ratings():
    conn=sqlite3.connect("movieRecommenderSystem.db")
    cur=conn.cursor()
    cur.execute("CREATE TABLE IF NOT EXISTS moviesGenre (id INTEGER PRIMARY KEY, movieId INTEGER, movieName text, genre text,rating REAL)")
    conn.commit()
    conn.close()
```

```python
def insert_movie_ratings(movieId,movieName,genre,rating):
    conn=sqlite3.connect("movieRecommenderSystem.db")
    cur=conn.cursor()
    cur.execute("INSERT INTO moviesGenre VALUES (NULL,?,?,?,?)",(movieId,movieName,genre,rating))
    conn.commit()
    conn.close()
```

```python
connect_movie_ratings()
```

```python
i=0
for itr in gen_list:
    i=i+1
    print(i)
    insert_movie_ratings(itr[0],itr[1],itr[2],itr[3])
```

```python
def view_userRegister():
    conn=sqlite3.connect("movieRecommenderSystem.db")
    cur=conn.cursor()
    cur.execute("SELECT * FROM moviesGenre")
    rows=cur.fetchall()
    conn.close()
    print(rows)
```
